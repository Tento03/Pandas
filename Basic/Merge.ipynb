```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
shp=[(1,'Luffy','Captain'),
     (2,'Zoro','Swordsman'),
     (3,'Sanji','Cook'),
     (4,'Nami','Navigator'),
     (5,'Ussop','sniper'),
     ]
```

```python
df=[(1,'nika'),(2,'none'),(3,'none'),(4,'none'),(5,'none')]
```

```python
df=pd.DataFrame(df,columns=['id','df'])
df
```

```python
shp=pd.DataFrame(shp,columns=['id','Name','Captain'])
shp
```

```python
dfmerge1=pd.merge(shp,df,on="id",how='inner')
dfmerge1
```

```python
dfmerge2=pd.merge(shp,df,on='id',how='outer')
dfmerge2
```

```python
dfmerge3=pd.merge(shp,df,on='id',how='left')
dfmerge3
```

```python
dfmerge4=pd.merge(shp,df,on='id',how='right')
dfmerge4
```

```python
dfmerge5=pd.merge(shp,df,on='id',how='outer',indicator=True)
dfmerge5
```
